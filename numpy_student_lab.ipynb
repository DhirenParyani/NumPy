```python
# TODO:
# Create random array of size 1000
rng = np.random.default_rng(10)
X = rng.standard_normal(1000)
X
```

```python
# TODO:
# Extract values greater than mean

# HINT:
# - Mean first
# - Boolean mask
mean = X.mean()
X_gt_mean = X[X > mean]

X_gt_mean
```

```python
# TODO:
# Replace negative values with 0 (no loops)
X_copy = X.copy()
X_copy[X_copy<0] = 0

X_copy

print('mean',mean)
print('above mean count:',X_gt_mean.size)
print('negative after replacing:',(X_copy < 0).sum())
```

Section 3 — Broadcasting

Task 3.1: Broadcasting Rules

```python
# TODO:
# Create A (1000, 50) and b (50,)
rng = np.random.default_rng(1)
A = rng.standard_normal((1000,50))
b = rng.standard_normal((50,))
b.shape
```

```python
# TODO:
# Add b to each row of A
A_plus = A + b
print("A",A)
print("b",b)
print("A_plus",A_plus)
A_plus.shape
# HINT:
# - No reshape required
```

```python
# TODO:
#one of the usecases of broadcasting is normalization
# Normalize each row of A
# Normalization = rescaling data so values are on a comparable scale.
# HINT:
# - Axis matters
# - Keep dimensions in mind
# there are different types of normalization, min max normalization and this is the standard normalization
# row - mean(row) / std(row)
row_mean = A.mean(axis=1,keepdims=True)
#keep dims make other column as 1 otherwise it would have been a 1D array with 1000 values
#for 1000 row we have the mean in single col
#row_mean.shape
#standard divation is square root of each value in the row - mean / total no of values (sq root of avg squared distance from mean)
row_std = A.std(axis=1,keepdims=True)
#row_std.shape
A_norm = (A-row_mean)/row_std
A_norm.shape
```

Section 4 — Vectorization vs Loops

Task 4.1: Loop vs Vectorized

```python
# TODO:
# Create large array X of size 1,000,000
rng = np.random.default_rng(2)
X = rng.standard_normal(1_000_000)
X.shape
```

```python
# TODO:
# Normalize using Python loop

# HINT:
# - Time it
#mean of all 1 million values
mu = X.mean()
#std of all 1 million values
sigma = X.std()
t0 = time.time()
out_loop = np.empty_like(X)
for i in range(X.shape[0]):
  out_loop[i] = (X[i]-mu)/sigma
time_taken = time.time() - t0
print(round(time_taken,4))


```

```python
# TODO:
# Normalize using vectorization
t1= time.time()
out_vec = (X - mu)/sigma
vec_time = time.time() - t1
print(round((time_taken)/max(vec_time,1e-12),1))
```

:Why is vectorization faster?
